### What are the most requested neighborhoods by AirBnb customers?

#### We will have to analyse the distribution of neighborhoods in the listings dataset.

```python
# Import necessary packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from time import time
```

```python
# Load the necessary listings dataset
data = pd.read_csv("C:/Users/user/Desktop/PYTHON/Udacity/seattle airbnb/listings.csv")
data.head()
```

```python
#load the necessar columns of the listings dataset
data = data[['neighbourhood', 'city', 'state']]
data
```

```python
data["city"].value_counts()
```

```python
data["state"].value_counts()
```

#### All the neighbourhood are in Seattle except one which contain some strange value

```python
data = data[data.city != "西雅图"]
data["city"].value_counts()
```

```python
# Get the 10 requested neighbourhood
neighbourhood_count = data['neighbourhood'].value_counts()
neighbourhood_count.head(10)
```

```python
# Percentage of the neighbhbourhood
df = pd.DataFrame({"Neighbourhood" : neighbourhood_count.index[:10],
                   "Required %" : 100 * neighbourhood_count.iloc[:10].values/ neighbourhood_count.sum()
                   })
df.head(10)
```

```python
# Bar plot
fig = plt.figure(figsize=(13, 10))
ax = fig.add_subplot(1, 1, 1, aspect='auto')
sns.barplot(x=df.Neighbourhood, y=df['Required %'])
# Set axis label properties
ax.set_xlabel('Neighborhood', weight='normal', size=20)
ax.set_ylabel('Percentage of requests % ', weight='normal', size=20)
plt.title('10 most requested neighborhoods', fontsize=20)
# Set tick label properties
ax.tick_params('x', labelsize=12, rotation=25)
ax.tick_params('y', labelsize=15)
plt.show()
```

#### Here you can see in descending order the proportion of requests related to each neighborhood in Seattle in 2015.
#### The clear leader in the Capitol Hill neighborhood with slightly more than 10% of the requests.
#### There is a wider gap between the first and the second place,
#### which is Ballard with slightly above 6% followed by Belltown and Minor neighborhoods with a narrow difference. 
#### In the next band, we have both Fremont and Wallingford neighborhoods with approximately 5 % of requests

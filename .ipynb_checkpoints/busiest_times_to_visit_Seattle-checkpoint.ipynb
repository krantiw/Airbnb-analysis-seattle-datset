### What are the busiest times of the year to visit Seattle? By how much do prices spike ?

### To answer the above question, lets we'll have to use calender and review datasets.
###### We'll require- 1.listing_id and date columns in the reviews dataset. 
######                        2.listing id, data and price in the calender dataset.

```python
# Import the necessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from time import time
```

```python
#loading the datasets
calendar = pd.read_csv("C:/Users/user/Desktop/PYTHON/Udacity/seattle airbnb/calendar.csv",parse_dates=['date'])
reviews = pd.read_csv("C:/Users/user/Desktop/PYTHON/Udacity/seattle airbnb/reviews.csv",parse_dates=['date'])
```

```python
#exploring the datasets
print(calendar.head())
print(reviews.head())
```

```python
#dropping the unwanted columns
calendar=calendar.drop(['available'], axis=1)
calendar.head()
```

```python
reviews= reviews.drop(["reviewer_id","reviewer_name", "comments", "id"], axis=1)
reviews.head()
```

```python
print(reviews.shape)
print(calendar.shape)
```

```python
# Get the date range in reviews dataset
reviews_freq = pd.DataFrame(reviews['date'].value_counts().values,
                        index=reviews['date'].value_counts().index,
                        columns=['rev_num'])
reviews_freq
```

```python
# resampling and suming the monthly data into yearly data
reviews_freq_year = reviews_freq.resample('A').sum()
reviews_freq_year
```

```python
#Get the range in calender data set
calendar_feq = pd.DataFrame(calendar["date"].value_counts().values,
                           index=calendar["date"].value_counts().index,
                           columns=["price_num"])
calendar_feq
```

```python
# resampling and suming the monthly data into yearly data
calendar_feq_year = calendar_feq.resample("A").sum()
calendar_feq_year
```

From the analyse done above we may notice :

- The reviews dataset contains information between 2009 and 2016. 
- The calendar dataset contains information between 2016 and 2017.
- The year 2015 contains the highest number of information recorded in the reviews dataset.
- The year 2016 contains the highest number of information recorded in the calendar dataset.

Therfore we will use the year 2015 to analyse the busiest times of the year to visit Seattle. And the year 2016 to analyse the price variation during the year.

### What are the busiest times of the year to visit Seattle ?

```python
# Selct the year 2015 of the reviews dataset since it contains the largest information
reviews_freq_2015 = reviews_freq.loc['2015']
reviews_freq_2015
```

```python
# The summation of reservations made per month in the year 2015
res_reviews_2015_month = reviews_freq_2015.resample('M').sum()
res_reviews_2015_month
```

```python
#Calculate the monthly percentage of reservations
res_reviews_2015_month['% rev'] = res_reviews_2015_month['rev_num']*100/res_reviews_2015_month[
    'rev_num'].sum()
res_reviews_2015_month
```

### Plot the Data

```python
# Bar plot
fig1 = plt.figure(figsize=(15, 10))
ax = fig1.add_subplot(1, 1, 1, aspect='auto')
sns.barplot(x=res_reviews_2015_month.index.month_name(), y=res_reviews_2015_month['% rev'])

# Set axis label properties
ax.set_xlabel('Month', weight='normal', size=20)
ax.set_ylabel('Reviews percentage %', weight='normal', size=20)
plt.title('Reviews percentage per month in 2015', fontsize=20)

# Set tick label properties
ax.tick_params('x', labelsize=15, rotation=35)
ax.tick_params('y', labelsize=15)
plt.show()
```

#### From the figure below, we can see that the busiest time is the summer period with around 15 % of reviews 
#### in ascending order to the end of the year with about 6% of reviews.


#### Contrary to this is,
#### the start of the year period with less than 4% which then ascends to the peak at the summers.
